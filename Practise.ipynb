```python
#conditional stmt

def school_year_group(syear):
  if syear.is_integer():
    if syear < 6:
      print("Primary school")
    elif syear > 6 and syear <=10:
      print("Secondary school")
    elif syear > 10 and syear <=12:
      print("High school")
    else:
      print("invalid year for school")
  else:
    print("Please enter a valid number")

school_year = float(input("Enter your school year: "))
school_year_group(school_year)
```

```python
def calc(n,operation="add"):
  if operation == "add":
    return n+n
  elif operation == "square":
    return n*n
  elif operation == "subtract":
    return n-n
  else:
    return "invaid"

my_num = int(input("Enter a number to calc: "))
op = input("Enter operation to calc: ")

print(calc(my_num,op))
```

```python
import sys
import time
import numpy as np
size = 1000000

l1 = range(size)
l2 = range(size)

np1 = np.arange(size)
np2 = np.arange(size)

start = time.time()
result = [(x+y) for x,y in zip(l1,l2)]
end = time.time()
elapsed = round((end-start)*1000,2)
res_size = sys.getsizeof(1) * len(result)
print("Time to process size {} using list is {} milliseconds.".format(res_size,elapsed))

start = time.time()
np3 = np1 + np2
end = time.time()
elapsed = round((end - start)/1000,2)
npsize = np3.size * np3.itemsize

print("Time to process size {} using Numpy is {} milliseconds.".format(npsize,elapsed))
```
